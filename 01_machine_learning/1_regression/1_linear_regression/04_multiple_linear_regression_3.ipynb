# statsmodels.api : sm

### stats module -> 1st column of x vector(x data) has to be 1 constant vector

```python
x=sm.add_constant(x)
x.head()
```

```python
y=df['medv']
y.head()
```

```python
df_ols=sm.OLS(y,x)
statistic=df_ols.fit()
statistic.summary()
```

# statsmodel.fomula.api : smf

### 장점 : R과 비슷한 직관성을 가지고 있음.

```python
import statsmodels.formula.api as smf
r = smf.ols('medv ~ crim+zn+indus',data=df).fit()
r.summary()
```
