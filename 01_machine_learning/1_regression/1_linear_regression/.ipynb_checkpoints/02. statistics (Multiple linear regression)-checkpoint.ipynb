# Varialbe selection in Multiple linear regression(by RMSE)

### model : y = b0 + b1x2 + b2x2+ b3x3 + e

```python
lm=LinearRegression()
```

```python
df2=df.iloc[:, :-1]
```

```python
from  itertools import combinations
combination = list(combinations(df2.columns, 3))
combination
```

## RMSE

```python
rmse_list=list()

for i in range(len(combination)):
    x=df2.loc[:, combination[i]]
    y=df['medv']
    
    lm.fit(x,y)
    y_hat=lm.predict(x)
    
    rmse=np.sqrt(((y-y_hat)**2).mean())
    rmse_list.append(rmse)
```

```python
rmse_list
```

# selected variables

```python
rmse_list.index(min(rmse_list))
```
