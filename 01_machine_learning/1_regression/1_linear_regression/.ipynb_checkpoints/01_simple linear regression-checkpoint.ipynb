```python
import numpy as np
import scipy as sp
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt
import math
import os
from sklearn.linear_model import LinearRegression
```

```python
os.getcwd()
os.chdir('/home/ktai17/Downloads/data/data3')
os.getcwd()
```

```python
df=pd.read_csv('data05_boston.csv')
df.head()
```

```python
x=df.loc[:, ['lstat']]
x.head()
```

```python
y=df.loc[:, ['medv']]
y.head()
```

# Simple Linear Regression

```python
lm=LinearRegression()
lm
```

```python
lm.fit(x,y)
print(lm.coef_, lm.intercept_)
```

#### i.e, the model is  
#### y=-0.95004935x + 34.55384088

# Prediction

```python
# linearly estimate
y_hat=lm.predict(x)
y_hat
```

# R square, MSE, RMSE

```python
r2=lm.score(x,y)
r2
```

```python
mse=((y-y_hat)**2).mean()
mse
```

```python
rmse = np.sqrt(((y-y_hat)**2).mean())
rmse
```

```python
# plotting

plt.title('All vs. Medv: %.2f' % r2)
plt.plot(x, y, 'bo')
plt.plot(x, y_hat, 'r')
plt.show()
```
