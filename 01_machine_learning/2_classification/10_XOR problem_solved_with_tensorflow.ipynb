```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
x_data = [[0,0], [1,0], [0,1], [1,1]]
y_data = [[0], [1], [1], [0]]
```

```python
X= tf.placeholder(tf.float32, shape=[None, 2])
Y= tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W1= tf.Variable(tf.random_normal([2, 3]), name='weight1')
b1= tf.Variable(tf.random_normal([3]), name='bias1')
W2= tf.Variable(tf.random_normal([3, 6]), name='weight2')
b2= tf.Variable(tf.random_normal([6]), name='bias2')
W3= tf.Variable(tf.random_normal([6, 1]), name='weight3')
b3= tf.Variable(tf.random_normal([1]), name='bias3')


model1= tf.add(tf.matmul(X, W1), b1) #linear regression
model1= tf.sigmoid(model1) # Logistic(sigmoid)
model2= tf.add(tf.matmul(model1, W2), b2) #linear regression
model2= tf.sigmoid(model2) # Logistic(sigmoid)
model3= tf.add(tf.matmul(model2, W3), b3) #linear regression
model3= tf.sigmoid(model3) # Logistic(sigmoid)

cost= tf.reduce_mean((-1)*Y*tf.log(model3)+(-1)*(1-Y)*tf.log(1-model3))
train=tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

prediction = tf.cast(model3>0.5, dtype=tf.float32)
accuracy= tf.reduce_mean(tf.cast(tf.equal(prediction, Y), dtype=tf.float32))
```

```python
sess= tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
## training

for step in range(100001):
    cost_val, _=sess.run([cost, train], feed_dict={X:x_data, Y:y_data})
    if step % 200 == 0:
        print(step, cost_val)
```

```python
'''
Answer
'''
h, c, a = sess.run([model3, prediction, accuracy], feed_dict={X:x_data, Y:y_data})
print("\nModel:\n", h, "\nCorrect:\n ", c, "\nAccuracy:\n ", a)
```
