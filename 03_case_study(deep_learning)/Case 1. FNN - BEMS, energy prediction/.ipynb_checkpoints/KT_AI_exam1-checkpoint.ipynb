```python
####################################################################
# Part #0: 기초 준비 사항
####################################################################

# Import packages
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import matplotlib.pyplot as plt

# Setting for tensorflow runs
# (1) 연산: 노드(점)와 엣지(선)로 구성된 graph 표현
# (2) 그래프: 노드(i.e. 작업(OP))는 0개 이상의 Tensor를 만들 수 있음.
# (3) 텐서: 다차원 배열(e.x. [x1, x2, x3, x4]--> 4차원 배열)
# (4) 실행: Session --> 그래프에서 작업을 CPU OR GPU 같은 Device에 배정하고, 실행을 위한 메서드를 제공함.

# 텐선플로우 구성
# (1) 구성단계(Construction phase): graph를 조립하는 단계
# (2) 실행단계(Execution phase): session을 이용하여 그래프의 작업을 실행하는 단계

# 난수 텐서
# (1) 텐서플로우는 상이한 분포를 가진 난수 텐서를 생성 --> 연산
# (2) seed 키워드: 그래프의 난수 시드값과 동시에 작동함.
# (3) set_random_seed: 그래프의 난수 시드를 설정하는 것으로, 모든 연산에 생성된 난수 시퀀스들이 세션간 반복이 가능하게 하기 위한 것임.
# 예제:
# tf.set_random_seed(1234)
# a = tf.random_uniform([1])
# b = tf.random_uniform([1])

# print("Session 1")
# with tf.Session() as sess1:
#   print(sess1.run(a))  # generates 'A1'
#   print(sess1.run(a))  # generates 'A2'
#   print(sess1.run(b))  # generates 'B1'
#   print(sess1.run(b))  # generates 'B2'

# print("Session 2")
# with tf.Session() as sess2:
#   print(sess2.run(a))  # generates 'A1'
#   print(sess2.run(a))  # generates 'A2'
#   print(sess2.run(b))  # generates 'B1'
#   print(sess2.run(b))  # generates 'B2'

# InteractiveSession 클래스
# Tensor.eval()과 Operation.run() 메소드 호출 
# 그러면 세션을 위해 변수를 미리 생성할 필요가 없습니다.

tf.set_random_seed(777)
sess = tf.InteractiveSession()
```

```python
####################################################################
# Part #1: 데이터 읽기
####################################################################
# Pandas 이용 --> reference: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_excel.html 
# ! pip install xlrd
Nor_data = pd.read_excel(r'.\FDD_raw_data.xlsx', sheetname='정상적인 데이터')
```

```python
####################################################################
# Part #2: 정상적인 데이터 추출
####################################################################

###################################################
# 총 Data 항목
###################################################
# T02 : 실외온도
# T03 : 실외상대습도
# T04 : 풍속
# T05 : 실외 확산 일사량
# T06 : 실외 직달 일사량
# T07 : 냉수출구온도
# T08 : 냉수출구유량
# T09 : 냉수입구온도
# T10 : 리턴팬 토출구 온도
# T11 : 리터팬 토출구 유량
# T12 : 공급팬 토출구 온도
# T13 : 공급팬 토출구 유량
# T14 : 외기 댐퍼 토출구 온도
# T15 : 외기 댐퍼 토출구 유량
###################################################

###################################################
# 주관적 판단으로 정상데이터 입력/출력 데이터 구분
###################################################

# 정상데이터 입력
px1 = np.array(Nor_data['T02'], dtype=np.float32)           # 실외온도
px2 = np.array(Nor_data['T08'], dtype=np.float32)           # 냉수출구유량
px3 = np.array(Nor_data['T10'], dtype=np.float32)           # 리턴팬 토출구 온도
px4 = np.array(Nor_data['T11'], dtype=np.float32)           # 리턴팬 토출구 유량
px5 = np.array(Nor_data['T13'], dtype=np.float32)           # 공급팬 토출구 유량
px6 = np.array(Nor_data['T14'], dtype=np.float32)           # 외기 댐퍼 토출구 온도
px7 = np.array(Nor_data['T15'], dtype=np.float32)           # 외기 댐퍼 토출구 유량



# 정상데이터 출력
py1 = np.array(Nor_data['T12'], dtype=np.float32)           # 공급팬 토출구 온도

# 정상데이터 입력 stack
x_pre = tf.stack([px1, px2, px3, px4, px5, px6, px7]).eval()
x_f = tf.transpose(x_pre).eval()

print(x_f)
print(x_f.shape)

# 정상데이터 출력 stack
y_pre = tf.stack([py1]).eval()
y_f = tf.transpose(y_pre).eval()
print(y_f)
print(y_f.shape)
```

```python
###################################################
# Part #3: 정상 데이터의 훈련/검증 구분
###################################################

# train/test split for Normal data --> 과적합 방지 중 하나임.
train_size = int(len(x_f) * 0.8)

trainX, testX = np.array(x_f[0:train_size]), np.array(x_f[train_size:len(x_f)])
trainY, testY = np.array(y_f[0:train_size]), np.array(y_f[train_size:len(y_f)])
```

```python
###################################################
# Part #4: 정규화 과정 - 표준편차 정규화 실시
###################################################

# Q1: Decimal scaling normlaization -> V'(i) = v(i)/(10^k)
# Q2: Min-Max normalization -> v'(i)=(v(i)-min[v(i)])/(max[v(i)]-min[v(i)])
# Q3: Standard deviation normalization -> v'(i)=(v(i)-mean[v])/sd[v]
# Q1-2 

# Normalization - train
Nor_trainX = tf.transpose(trainX).eval()
Nor_trainY = tf.transpose(trainY).eval()

Normal_x1_Tr =(Nor_trainX[0] - Nor_trainX[0].mean()) / Nor_trainX[0].std()
Normal_x2_Tr =(Nor_trainX[1] - Nor_trainX[1].mean()) / Nor_trainX[1].std()
Normal_x3_Tr =(Nor_trainX[2] - Nor_trainX[2].mean()) / Nor_trainX[2].std()
Normal_x4_Tr =(Nor_trainX[3] - Nor_trainX[3].mean()) / Nor_trainX[3].std()
Normal_x5_Tr =(Nor_trainX[4] - Nor_trainX[4].mean()) / Nor_trainX[4].std()
Normal_x6_Tr =(Nor_trainX[5] - Nor_trainX[5].mean()) / Nor_trainX[5].std()
Normal_x7_Tr =(Nor_trainX[6] - Nor_trainX[6].mean()) / Nor_trainX[6].std()

Normal_y1_Tr =(Nor_trainY[0] - Nor_trainY[0].mean()) / Nor_trainY[0].std()

Mean_Y_tr = Nor_trainY[0].mean()
Std_Y_tr = Nor_trainY[0].std()

x_pre_Tr = tf.stack([Normal_x1_Tr, Normal_x2_Tr, Normal_x3_Tr, Normal_x4_Tr, Normal_x5_Tr, Normal_x6_Tr, Normal_x7_Tr]).eval()
x_f_Tr = tf.transpose(x_pre_Tr).eval()

y_pre_Tr = tf.stack([Normal_y1_Tr]).eval()
y_f_Tr = tf.transpose(y_pre_Tr).eval()

# Normalization - test
Nor_testX = tf.transpose(testX).eval()
Nor_testY = tf.transpose(testY).eval()

Normal_x1_Te =(Nor_testX[0] - Nor_testX[0].mean()) / Nor_testX[0].std()
Normal_x2_Te =(Nor_testX[1] - Nor_testX[1].mean()) / Nor_testX[1].std()
Normal_x3_Te =(Nor_testX[2] - Nor_testX[2].mean()) / Nor_testX[2].std()
Normal_x4_Te =(Nor_testX[3] - Nor_testX[3].mean()) / Nor_testX[3].std()
Normal_x5_Te =(Nor_testX[4] - Nor_testX[4].mean()) / Nor_testX[4].std()
Normal_x6_Te =(Nor_testX[5] - Nor_testX[5].mean()) / Nor_testX[5].std()
Normal_x7_Te =(Nor_testX[6] - Nor_testX[6].mean()) / Nor_testX[6].std()

Normal_y1_Te =(Nor_testY[0] - Nor_testY[0].mean()) / Nor_testY[0].std()

Mean_Y_te = Nor_testY[0].mean()
Std_Y_te = Nor_testY[0].std()

x_pre_Te = tf.stack([Normal_x1_Te, Normal_x2_Te, Normal_x3_Te, Normal_x4_Te, Normal_x5_Te, Normal_x6_Te, Normal_x7_Te]).eval()
x_f_Te = tf.transpose(x_pre_Te).eval()

y_pre_Te = tf.stack([Normal_y1_Te]).eval()
y_f_Te = tf.transpose(y_pre_Te).eval()
```

```python
####################################################################
# Part #5: 딥러닝 실시
####################################################################

# Placeholder: training set을 저장하는 변수

X = tf.placeholder(tf.float32, shape=[None, 7])
Y = tf.placeholder(tf.float32, shape=[None, 1])
#keep_prob = tf.placeholder(tf.float32)


# Variable: 최적화 대상 파라미터 -> Weight / Bias
# Q1. Hidden Layer와 neuron 변경
# Q2. 활성화 함수 변경: Sigmoid, tanh, relu, Leaky relu, ...
# Q3. 절단율 적용: Dropout

W1 = tf.get_variable('W1', shape=[7, 5], initializer=tf.contrib.layers.xavier_initializer())
b1 = tf.Variable(tf.random_normal([5]))
L1 = tf.nn.relu(tf.matmul(X, W1) + b1)
#L1 = tf.nn.dropout(L1, keep_prob=0.8)

W2 = tf.get_variable('W2', shape=[5, 1], initializer=tf.contrib.layers.xavier_initializer())
b2 = tf.Variable(tf.random_normal([1]))
hypothesis = tf.matmul(L1, W2) + b2

# cost/loss function
cost = tf.reduce_sum(tf.square(hypothesis - Y))  # sum of the squares

# optimizer
# Q4. Gradient decent?, Momentum?, AdaGrad?, Adam?
# 학습률: 0.01, 0.001, 0.0001, 0.00001
learning_rate = 0.001
optimizer = tf.train.AdamOptimizer(learning_rate)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()

# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

# Learning
# Q5. epochs 변경  
epochs = 10001

for step in range(epochs):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict={X: x_f_Tr, Y: y_f_Tr})
    #cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict={X: x_f_Tr, Y: y_f_Tr, keep_prob: 0.8})
    
    if step == 0:
        save_cost_val = []
    else:
        save_cost_val = save_cost_val

    save_cost_val.append(cost_val)
    
    if step % 1000 == 0:
        print(step, "Loss: ", cost_val)
        
y_pred = sess.run(hypothesis, feed_dict={X: x_f_Te})
#y_pred = sess.run(hypothesis, feed_dict={X: x_f_Te, keep_prob: 1})

```

```python
####################################################################
# Part #6: Cost/loss function과 Validation 확인
####################################################################

# Figure 1: Cost/Loss function 
plt.figure(1)
plt.subplot(211)
plt.plot(save_cost_val, 'r-')
plt.xlabel('Step size')
plt.ylabel('Loss')

plt.subplot(212)
plt.plot(pd.Series(np.ravel(y_f_Te*Std_Y_te+Mean_Y_te)), "r-", markersize=2, label="Actual")
plt.plot(pd.Series(np.ravel(y_pred*Std_Y_te+Mean_Y_te)), "b--", markersize=2, label="Forecast")
plt.legend(loc="upper right")
plt.xlabel("Time(10min)")
plt.ylabel('Supply Air Temperaure[C]')

plt.show()
```

```python
####################################################################
# Part #7: 비정상적 데이터 추출
####################################################################

Anor_data = pd.read_excel(r'.\FDD_raw_data.xlsx', sheetname='비정상적인 데이터')

# 비정상적 데이터 입력변수
tx1 = np.array(Anor_data['T02'], dtype=np.float32)           # 실외온도
tx2 = np.array(Anor_data['T08'], dtype=np.float32)           # 냉수출구유량
tx3 = np.array(Anor_data['T10'], dtype=np.float32)           # 리턴팬 토출구 온도
tx4 = np.array(Anor_data['T11'], dtype=np.float32)           # 리턴팬 토출구 유량
tx5 = np.array(Anor_data['T13'], dtype=np.float32)           # 공급팬 토출구 유량
tx6 = np.array(Anor_data['T14'], dtype=np.float32)           # 외기 댐퍼 토출구 온도
tx7 = np.array(Anor_data['T15'], dtype=np.float32)           # 외기 댐퍼 토출구 유량

# 비정상적 데이터 출력변수
ty1 = np.array(Anor_data['T12'], dtype=np.float32)           # 공급팬 토출구 온도

# 시뮬레이션 간격: 시간 * 10분 간격 (=60/10)
Time_period = 24 * 6

# 임계값 설정
Threshold = 30

for iter in range(int(len(Anor_data)/Time_period)):

    ConvX1 = tx1[iter*Time_period:iter*Time_period+(Time_period)]
    ConvX2 = tx2[iter*Time_period:iter*Time_period+(Time_period)]
    ConvX3 = tx3[iter*Time_period:iter*Time_period+(Time_period)]
    ConvX4 = tx4[iter*Time_period:iter*Time_period+(Time_period)]
    ConvX5 = tx5[iter*Time_period:iter*Time_period+(Time_period)]
    ConvX6 = tx6[iter*Time_period:iter*Time_period+(Time_period)]
    ConvX7 = tx7[iter*Time_period:iter*Time_period+(Time_period)]

    ConvY1 = ty1[iter*Time_period:iter*Time_period+(Time_period)]

    # 비정상적인 데이터 정규화
    Abnormal_tx1 = (ConvX1 - Nor_testX[0].mean()) / Nor_testX[0].std()
    Abnormal_tx2 = (ConvX2 - Nor_testX[1].mean()) / Nor_testX[1].std()
    Abnormal_tx3 = (ConvX3 - Nor_testX[2].mean()) / Nor_testX[2].std()
    Abnormal_tx4 = (ConvX4 - Nor_testX[3].mean()) / Nor_testX[3].std()
    Abnormal_tx5 = (ConvX5 - Nor_testX[4].mean()) / Nor_testX[4].std()
    Abnormal_tx6 = (ConvX6 - Nor_testX[5].mean()) / Nor_testX[5].std()
    Abnormal_tx7 = (ConvX7 - Nor_testX[6].mean()) / Nor_testX[6].std()

    Abnormal_ty1 = (ConvY1 - Nor_testY[0].mean()) / Nor_testY[0].std()

    xt_pre = tf.stack([Abnormal_tx1, Abnormal_tx2, Abnormal_tx3, Abnormal_tx4, Abnormal_tx5, Abnormal_tx6, Abnormal_tx7]).eval()
    xt_f = tf.transpose(xt_pre).eval()

    yt_pre = tf.stack([Abnormal_ty1]).eval()
    yt_f = tf.transpose(yt_pre).eval()
    
    Pred_val = sess.run(hypothesis, feed_dict={X: xt_f})

    # Predicted vs. Measured
    Error_Test_result = tf.reduce_sum(tf.abs((Pred_val*Std_Y_te+Mean_Y_te) - (yt_f*Std_Y_te+Mean_Y_te))).eval()
    
    # Error 비교
    Error = Threshold - Error_Test_result
    
    # Error 양수이면 Fault 없음., Error 음수이면 Fault 있음.
    if Error >= 0:
        Fault_Alarm = 0
    else:
        Fault_Alarm = 1
    
    print('Iteration numb is', iter + 1)
    print('Error result is', Error_Test_result)
    print('Error is', Error)
    print('Fault Alarm is', Fault_Alarm)

    if iter == 0:
        save_Error_result = []
        save_Error_dev = []
        save_Fault = []
    else:
        save_Error_result = save_Error_result
        save_Error_dev = save_Error_dev
        save_Fault = save_Fault

    save_Error_result.append(Error_Test_result)
    save_Error_dev.append(Error)
    save_Fault.append(Fault_Alarm)
```

```python
####################################################################
# Part #8: 오류 검출 그래프
####################################################################

plt.figure(1)
plt.subplot(211)
plt.plot(save_Error_dev, 'r-', markersize=2)
plt.xlabel('Time(10 min)')
plt.ylabel('Error')
plt.axis([0, 62, -20, 10])

plt.subplot(212)
plt.plot(save_Fault, 'b-', markersize=2)
plt.xlabel('Time(10 min)')
plt.ylabel('Fault Alarm')
plt.axis([0, 62, -0.1, 1.1])

plt.show()


```
